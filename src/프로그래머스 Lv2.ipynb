```java
// 타겟 넘버

class Solution {
    
    public int search(int[] numbers, int target, int result, int r) {
        if(r == numbers.length) {
            if(result == target) return 1;
            return 0;
        }
        return search(numbers, target, result + numbers[r], r+1) +
            search(numbers, target, result - numbers[r], r+1);
    }
    
    public int solution(int[] numbers, int target) {
        int answer = 0;
        answer = search(numbers, target, 0, 0);
        return answer;
    }
}
```

```java
// 큰 수 만들기

import java.util.*;

class Solution {
    public String solution(String number, int k) {
        String answer = "";
        boolean[] on = new boolean[number.length()];    // 지운거는 false
        Arrays.fill(on, true);
        for(int i = 0; i < number.length()-1;) {
            if(!on[i]) {
                i++;
                continue;
            }
            int j;
            for(j = i+1; !on[j]; j++);
            if(number.charAt(i)-'0' < number.charAt(j)-'0') {
                on[i] = false;
                k--;
                if(i == 0) i++;
                else {
                    for(int m = i; m >= 0; m--) {
                        if(on[m]) {
                            i = m;
                            break;
                        }
                    }
                }
            }
            else i++;
            if(k == 0) break;
        }
        if(k > 0) on[number.length()-1] = false;
        StringBuilder sb = new StringBuilder();
        for(int i = 0; i < number.length(); i++) {
            if(on[i]) sb.append(number.charAt(i));
        }
        answer = sb.toString();
        return answer;
    }
}
```

```java
// 프린터

import java.util.*;

class Solution {
    
    class Document implements Comparable<Document> {
        int id, priority;
        public Document(int id, int p) {
            this.id = id;
            this.priority = p;
        }
        @Override
        public int compareTo(Document o) {
            return Integer.compare(o.priority, this.priority);
        }
    }
    public int solution(int[] priorities, int location) {
        int answer = 0;
        PriorityQueue<Document> pq = new PriorityQueue<>();
        Queue<Document> queue = new LinkedList<>();
        for(int i = 0; i < priorities.length; i++) {
            Document d = new Document(i, priorities[i]);
            pq.add(d);
            queue.add(d);
        }
        int order = 0;
        while(!queue.isEmpty()) {
            Document d = queue.remove();
            if(pq.peek().priority == d.priority) {
                order++;
                if(d.id == location) return order;
                pq.remove();
            }
            else queue.add(d);
        }
        return answer;
    }
}
```

```java
// 구명보트

import java.util.*;

class Solution {
    public int solution(int[] people, int limit) {
        int answer = 0;
        PriorityQueue<Integer> pq = new PriorityQueue<>();
        for(int p : people) pq.add(p);
        int[] weight = new int[people.length];
        for(int i = 0; i < people.length; i++) weight[i] = pq.remove();
        int left = 0, right = people.length-1;
        while(left <= right) {
            if(weight[right] + weight[left] <= limit) {
                left++;
                right--;
            }
            else right--;
            answer++;
        }
        return answer;
    }
}
```

```java
// 다리를 지나는 트럭

import java.util.*;

class Solution {
    public int solution(int bridge_length, int weight, int[] truck_weights) {
        int answer = 0;
        Queue<Integer> queue = new LinkedList<>();
        int time = 1;
        int t = 0;
        int sum = truck_weights[t];
        queue.add(truck_weights[t++]);
        int passed = 0;
        while(passed < truck_weights.length) {
            time++;
            // 도착했으면 나오기
            if(queue.size() == bridge_length) {
                int truck = queue.remove();
                if(truck > 0) {
                    sum -= truck;
                    passed++;
                }
            }
            // 트럭 올라갈 수 있으면 올라가기
            if(t < truck_weights.length) {
                if(sum + truck_weights[t] <= weight) {
                    queue.add(truck_weights[t]);
                    sum += truck_weights[t++];
                }else queue.add(0);
            } else { // 모든 트럭이 올라간 경우
                if(passed == truck_weights.length) queue.remove();
                else queue.add(0);
            } 
        } 
        answer = time;
        return answer;
    }
}
```

```java
// 카펫

class Solution {
    public int[] solution(int brown, int yellow) {
        /*
        1 24 / 2 12 / 3 8 / 4 6
        (1+24+2)*2 = 27*2 = 54
        (2+12+2)*2 = 16*2 = 32
        (3+8+2)*2 = 13*2 = 26
        (4+6+2)*2 = 12*2 = 24
        */
        int width = 0, height = 0;
        for(int i = 1; i <= Math.sqrt(yellow); i++) {
            if(yellow % i == 0) {
                width = yellow / i;
                height = i;
                if((width + height + 2) * 2 == brown) break;
            }
        }
        int[] answer = {width+2, height+2};
        return answer;
    }
}
```

```java
// 소수 찾기

import java.util.*;

class Solution {
    
    static HashSet<Integer> set;
    
    public boolean isPrime(int n) {
        if(n <= 1) return false;
        if(n == 2) return true;
        for(int i = 2; i <= Math.sqrt(n); i++) {
            if(n%i == 0) return false;
        }
        return true;
    }
    
    public void permutation(String str, String sub, int n, int r, boolean[] visited) {
        if (r == n) {
            set.add(Integer.parseInt(sub));
            return;
        }
        for(int i = 0; i < n; i++) {
            if(!visited[i]) {
                visited[i] = true;
                permutation(str, sub+str.charAt(i), n, r+1, visited);
                visited[i] = false;
            }
        }
    }
    
    public void combination(String numbers, String sub, int start, int n, int r) {
        if(r == n) {
            permutation(sub, "", n, 0, new boolean[n]);
            return;
        }
        for(int i = start; i < numbers.length(); i++) combination(numbers, sub + numbers.charAt(i), i+1, n, r+1);
    }
    
    public int solution(String numbers) {
        int answer = 0;
        set = new HashSet<>();
        for(int i = 1; i <= numbers.length(); i++) combination(numbers, "", 0, i, 0);
        Iterator iter = set.iterator();
        while(iter.hasNext()) {
            if(isPrime(Integer.parseInt(iter.next().toString()))) answer++;
        }
        return answer;
    }
}
```

```java
// 예상 대진표

import java.util.*;

class Solution
{
    public int solution(int n, int a, int b)
    {
        int answer = 0;
        int[] tree = new int[n*2];
        for(int i = n; i < n*2; i++) tree[i] = i-n+1;
        int start = n;
        int cnt = 0;
        int temp = Math.max(a,b);
        a = Math.min(a,b);
        b = temp;
        boolean meet = false;
        while(start > 1) {
            cnt++;
            for(int i = start; i < start*2; i += 2) {
                if(tree[i] == a && tree[i+1] == b) {
                    meet = true;
                    break;
                }
                if(tree[i+1] == a || tree[i+1] == b) tree[i/2] = tree[i+1];
                else tree[i/2] = tree[i];
            }
            if(meet) break;
            start /= 2;
        }
        answer = cnt;
        return answer;
    }
}
```

```java
// 게임 맵 최단거리 - 효율성 실패

import java.util.*;

class Solution {
    
    int[] dr = {-1, 0, 1, 0};
    int[] dc = {0, 1, 0, -1};
    int n, m;
    
    class Pos {
        int i, j, cnt;
        public Pos(int i, int j, int cnt) {
            this.i = i;
            this.j = j;
            this.cnt = cnt;
        }
    }
    
    public int bfs(int[][] maps) {
        Queue<Pos> queue = new LinkedList<>();
        boolean[][] visited = new boolean[n][m];
        queue.add(new Pos(0,0,1));
        visited[0][0] = true;
        while(!queue.isEmpty()) {
            Pos p = queue.remove();
            if(p.i == n-1 && p.j == m-1) return p.cnt;
            for(int i = 0; i < 4; i++) {
                int ni = p.i + dr[i];
                int nj = p.j + dc[i];
                if(0 <= ni && ni < n && 0 <= nj && nj < m) {
                    if(!visited[ni][nj] && maps[ni][nj] == 1) {
                        visited[ni][nj] = true;
                        queue.add(new Pos(ni, nj, p.cnt+1));
                    }
                } 
            }
        }
        return -1;   
    }
    
    public int solution(int[][] maps) {
        n = maps.length;
        m = maps[0].length;
        return bfs(maps);
    }
}
```

```java
// 배달

import java.util.*;

class Solution {
    
    class Edge implements Comparable<Edge> {
        int v, weight;
        
        public Edge(int v, int weight) {
            this.v = v;
            this.weight = weight;
        }
        
        @Override
        public int compareTo(Edge o) {
            return Integer.compare(this.weight, o.weight);
        }
    }
    
    public int solution(int N, int[][] road, int K) {
        int answer = 1;
        ArrayList<Edge>[] G = new ArrayList[N+1];
        for(int i = 1; i <= N; i++) G[i] = new ArrayList<Edge>();
        for (int[] r : road) {
            G[r[0]].add(new Edge(r[1], r[2]));
            G[r[1]].add(new Edge(r[0], r[2]));
        }
        int[] dist = new int[N+1];
        boolean[] visited = new boolean[N+1];
        Arrays.fill(dist, Integer.MAX_VALUE);
        dist[1] = 0;
        PriorityQueue<Edge> pq = new PriorityQueue<>();
        pq.add(new Edge(1, 0));
        while(!pq.isEmpty()) {
            Edge e = pq.remove();
            if(visited[e.v]) continue;
            visited[e.v] = true;
            for(Edge ne : G[e.v]) {
                if(dist[e.v] + ne.weight < dist[ne.v]) dist[ne.v] = dist[e.v] + ne.weight;
                if(!visited[ne.v]) pq.add(new Edge(ne.v, dist[ne.v]));
            }
        }
        for(int i = 2; i <= N; i++) {
            if(dist[i] <= K) answer++;
        }

        return answer;
    }
}
```

```java
// 괄호 회전하기

import java.util.*;

class Solution {
    
    public boolean isRight(String s) {
        Stack<Character> stack = new Stack<>();
        for(int i = 0; i < s.length(); i++) {
            char c = s.charAt(i);
            if(c == '[' || c == '{' || c == '(') stack.push(c);
            else {
                if(stack.isEmpty()) return false;
                switch(c) {
                    case ']' :
                        if(stack.peek() == '[') stack.pop();
                        else return false;
                        break;
                    case '}' :
                        if(stack.peek() == '{') stack.pop();
                        else return false;
                        break;
                    case ')' :
                        if(stack.peek() == '(') stack.pop();
                        else return false;
                        break;
                }
            }
        }
        return true;
    }
    
    public int solution(String s) {
        if(s.length() % 2 != 0) return 0;
        int answer = 0;
        Queue<Character> queue = new LinkedList<>();
        for(int i = 0; i < s.length(); i++) queue.add(s.charAt(i));
        if(isRight(queue.toString().replaceAll(", ","").substring(1, s.length()+1))) answer++;
        for(int i = 1; i < s.length(); i++) {
            queue.add(queue.remove());
            if(isRight(queue.toString().replaceAll(", ","").substring(1, s.length()+1))) answer++;
        }
        return answer;
    }
}
```

```java
// 스킬트리

import java.util.*;

class Solution {
    public int solution(String skill, String[] skill_trees) {
        int answer = 0;
        for(String tree : skill_trees) {
            Queue<Character> queue = new LinkedList<>();
            boolean fail = false;
            for(int i = 0; i < skill.length(); i++) queue.add(skill.charAt(i));
            for(int i = 0; i < tree.length(); i++) {
                if(skill.contains(tree.charAt(i) + "")) {
                    if(queue.isEmpty()) {
                        fail = true;
                        break;
                    }
                    else {
                        if(tree.charAt(i) == queue.peek()) queue.remove();
                        else {
                            fail = true;
                            break;
                        }
                    }
                }
            }
            if(!fail) answer++;
        }
        return answer;
    }
}
```

```java
// 방문 길이

import java.util.*;

class Solution {
    
    class Dot {
        boolean U, R, D, L;    
        public Dot() {
            this.U = this.R = this.D = this.L = false;
        }
    }
    
    public int solution(String dirs) {
        int answer = 0;
        Dot[][] map = new Dot[11][11];
        for(int i = 0; i <= 10; i++) {
            for(int j = 0; j <= 10; j++) map[i][j] = new Dot();
        }
        int duplicate = 0, move = 0;
        int x = 5, y = 5;
        for(int i = 0; i < dirs.length(); i++) {
            char c = dirs.charAt(i);
            switch(c) {
                case 'U' :
                    if(y+1 <= 10) {
                        if(map[x][y].U) duplicate++;
                        else map[x][y].U = map[x][y+1].D = true;
                        y++;
                        move++;
                    }
                    break;
                case 'R' :
                    if(x+1 <= 10) {
                        if(map[x][y].R) duplicate++;
                        else map[x][y].R = map[x+1][y].L = true;
                        x++;
                        move++;
                    }
                    break;
                case 'D' :
                    if(y-1 >= 0) {
                        if(map[x][y].D) duplicate++;
                        else map[x][y].D = map[x][y-1].U = true;
                        y--;
                        move++;
                    }
                    break;
                case 'L' :
                    if(x-1 >= 0) {
                        if(map[x][y].L) duplicate++;
                        else map[x][y].L = map[x-1][y].R = true;
                        x--;
                        move++;
                    }
                    break;
            }
        }
        answer = move - duplicate;
        return answer;
    }
}
```

```java
// 전력망을 둘로 나누기

import java.util.*;

class Solution {
    int cnt;
    
    public void dfs(ArrayList[] tree, int n, boolean[] visited, int except) {
        cnt++;
        visited[n] = true;
        for(int node : (ArrayList<Integer>) tree[n]) {
            if(!visited[node] && node != except) dfs(tree, node, visited, except);
        }
    }
    
    public int solution(int n, int[][] wires) {
        int answer = -1;
        ArrayList<Integer>[] tree = new ArrayList[n+1];
        for(int i = 1; i <= n; i++) tree[i] = new ArrayList<Integer>();
        for(int[] wire : wires) {
            tree[wire[0]].add(wire[1]);
            tree[wire[1]].add(wire[0]);
        }
    
        int min = n;
        for(int[] wire : wires) {
            cnt = 0;
            dfs(tree, wire[0], new boolean[n+1], wire[1]);
            min = Math.min(min, Math.abs(cnt-n+cnt));
        }
        answer = min;
        return answer;
    }
}
```

```java
// 피로도

import java.util.*;

class Solution {
    
    int max;
    
    public void permutation(int[][] arr, int[] sub, int n, int r, boolean[] visited, int hp) {
        if(r == n) {
            max = r;
            return;
        }
        for(int i = 0; i < n; i++) {
            if(!visited[i]) {
                if(hp >= arr[i][0]) {
                    visited[i] = true;
                    sub[r] = i;
                    permutation(arr, sub, n, r+1, visited, hp-arr[i][1]);
                    visited[i] = false;
                }
                else max = Math.max(max, r);
            }
        }
    }
    
    public int solution(int k, int[][] dungeons) {
        int n = dungeons.length;
        max = 0;
        permutation(dungeons, new int[n], n, 0, new boolean[n], k);
        return max;
    }
}
```

```java
// N개의 최소공배수
import java.util.*;

class Solution {

    public int GCD(int a, int b){   // a가 큰 것
        if(b == 0) return a;
        else return GCD(b, a % b);
    }

    public int LCM(int a, int b) {
        int gcd = GCD(a, b);
        return a * b / gcd;
    }

    public int solution(int[] arr) {
        int answer = 0;
        PriorityQueue<Integer> pq = new PriorityQueue<>(Collections.reverseOrder());
        for(int a : arr) pq.add(a);
        while(pq.size() >= 2) {
            int c = LCM(pq.remove(), pq.remove());
            pq.add(c);
        }
        answer = pq.remove();
        return answer;
    }
}
```

```java
// 행렬의 곱셈
class Solution {
    public int[][] solution(int[][] arr1, int[][] arr2) {
        int row1 = arr1.length;
        int col1row2 = arr2.length;
        int col2 = arr2[0].length;
        int[][] answer = new int[row1][col2];
        for(int i = 0; i < row1; i++) {
            for(int j = 0; j < col2; j++) {
                for(int k = 0; k < col1row2; k++) answer[i][j] += (arr1[i][k]*arr2[k][j]);
            }
        }
        return answer;
    }
}
```
