```java
// 타겟 넘버

class Solution {
    
    public int search(int[] numbers, int target, int result, int r) {
        if(r == numbers.length) {
            if(result == target) return 1;
            return 0;
        }
        return search(numbers, target, result + numbers[r], r+1) +
            search(numbers, target, result - numbers[r], r+1);
    }
    
    public int solution(int[] numbers, int target) {
        int answer = 0;
        answer = search(numbers, target, 0, 0);
        return answer;
    }
}
```

```java
// 짝지어 제거하기 - 효율성 2개 실패

import java.util.*;

class Solution {
    public int solution(String s)
    {
        int[] x = new int[26];
        for(int i = 0; i < s.length(); i++) x[(int)s.charAt(i)-'a']++;
        for(int i = 0; i < 26; i++) if(x[i]%2 != 0) return 0;
        StringBuilder sb = new StringBuilder();
        while(true) {
            String temp = s;
            int i;
            for(i = 0; i < s.length()-1;) {
                if(s.charAt(i) == s.charAt(i+1)) i+=2;
                else {
                    sb.append(s.charAt(i));
                    i++;
                }
            }
            if(i == s.length()-1) sb.append(s.charAt(i));
            s = sb.toString();
            if(s.equals("")) return 1;
            if(s.equals(temp)) return 0;
            sb.setLength(0);
        }
    }
}
```

```java
// 큰 수 만들기

import java.util.*;

class Solution {
    public String solution(String number, int k) {
        String answer = "";
        boolean[] on = new boolean[number.length()];    // 지운거는 false
        Arrays.fill(on, true);
        for(int i = 0; i < number.length()-1;) {
            if(!on[i]) {
                i++;
                continue;
            }
            int j;
            for(j = i+1; !on[j]; j++);
            if(number.charAt(i)-'0' < number.charAt(j)-'0') {
                on[i] = false;
                k--;
                if(i == 0) i++;
                else {
                    for(int m = i; m >= 0; m--) {
                        if(on[m]) {
                            i = m;
                            break;
                        }
                    }
                }
            }
            else i++;
            if(k == 0) break;
        }
        if(k > 0) on[number.length()-1] = false;
        StringBuilder sb = new StringBuilder();
        for(int i = 0; i < number.length(); i++) {
            if(on[i]) sb.append(number.charAt(i));
        }
        answer = sb.toString();
        return answer;
    }
}
```

```java
// 전화번호 목록 - 효율성 2개 실패

import java.util.*;

class Solution {
    public boolean solution(String[] phone_book) {
        HashMap<Character, ArrayList<String>> map = new HashMap<>();
        for(String p : phone_book) {
            Character c = p.charAt(0);
            if(map.get(c) == null) {
                ArrayList<String> list = new ArrayList<>();
                list.add(p);
                map.put(c, list);
            }
            else {
                for(String phone : map.get(c)) {
                    if(phone.length() < p.length()) {
                        if(p.substring(0, phone.length()).equals(phone)) return false;
                    } else {
                        if(phone.substring(0, p.length()).equals(p)) return false;
                    }
                }
                map.get(c).add(p);
            }
        }
        return true;
    }
}
```

```java
// 프린터

import java.util.*;

class Solution {
    
    class Document implements Comparable<Document> {
        int id, priority;
        public Document(int id, int p) {
            this.id = id;
            this.priority = p;
        }
        @Override
        public int compareTo(Document o) {
            return Integer.compare(o.priority, this.priority);
        }
    }
    public int solution(int[] priorities, int location) {
        int answer = 0;
        PriorityQueue<Document> pq = new PriorityQueue<>();
        Queue<Document> queue = new LinkedList<>();
        for(int i = 0; i < priorities.length; i++) {
            Document d = new Document(i, priorities[i]);
            pq.add(d);
            queue.add(d);
        }
        int order = 0;
        while(!queue.isEmpty()) {
            Document d = queue.remove();
            if(pq.peek().priority == d.priority) {
                order++;
                if(d.id == location) return order;
                pq.remove();
            }
            else queue.add(d);
        }
        return answer;
    }
}
```

```java
// 구명보트

import java.util.*;

class Solution {
    public int solution(int[] people, int limit) {
        int answer = 0;
        PriorityQueue<Integer> pq = new PriorityQueue<>();
        for(int p : people) pq.add(p);
        int[] weight = new int[people.length];
        for(int i = 0; i < people.length; i++) weight[i] = pq.remove();
        int left = 0, right = people.length-1;
        while(left <= right) {
            if(weight[right] + weight[left] <= limit) {
                left++;
                right--;
            }
            else right--;
            answer++;
        }
        return answer;
    }
}
```

```java
// 다리를 지나는 트럭

import java.util.*;

class Solution {
    public int solution(int bridge_length, int weight, int[] truck_weights) {
        int answer = 0;
        Queue<Integer> queue = new LinkedList<>();
        int time = 1;
        int t = 0;
        int sum = truck_weights[t];
        queue.add(truck_weights[t++]);
        int passed = 0;
        while(passed < truck_weights.length) {
            time++;
            // 도착했으면 나오기
            if(queue.size() == bridge_length) {
                int truck = queue.remove();
                if(truck > 0) {
                    sum -= truck;
                    passed++;
                }
            }
            // 트럭 올라갈 수 있으면 올라가기
            if(t < truck_weights.length) {
                if(sum + truck_weights[t] <= weight) {
                    queue.add(truck_weights[t]);
                    sum += truck_weights[t++];
                }else queue.add(0);
            } else { // 모든 트럭이 올라간 경우
                if(passed == truck_weights.length) queue.remove();
                else queue.add(0);
            } 
        } 
        answer = time;
        return answer;
    }
}
```

```java
// 카펫

class Solution {
    public int[] solution(int brown, int yellow) {
        /*
        1 24 / 2 12 / 3 8 / 4 6
        (1+24+2)*2 = 27*2 = 54
        (2+12+2)*2 = 16*2 = 32
        (3+8+2)*2 = 13*2 = 26
        (4+6+2)*2 = 12*2 = 24
        */
        int width = 0, height = 0;
        for(int i = 1; i <= Math.sqrt(yellow); i++) {
            if(yellow % i == 0) {
                width = yellow / i;
                height = i;
                if((width + height + 2) * 2 == brown) break;
            }
        }
        int[] answer = {width+2, height+2};
        return answer;
    }
}
```

```java
// JadenCase 문자열 만들기

import java.util.*;

class Solution {
    public String solution(String s) {
        String answer = "";
        StringBuilder sb = new StringBuilder();
        sb.append(Character.toUpperCase(s.charAt(0)));
        for(int i = 1; i < s.length(); i++) {
            if(s.charAt(i-1) == ' ') sb.append(Character.toUpperCase(s.charAt(i)));
            else sb.append(Character.toLowerCase(s.charAt(i)));
        }
        answer = sb.toString();
        return answer;
    }
}
```

```java
// 영어 끝말잇기

import java.util.*;

class Solution {
    public int[] solution(int n, String[] words) {
        int[] answer = new int[2];
        HashSet<String> set = new HashSet<>();
        for(int i = 1; i <= words.length; i++) {
            // 말한 사람 : i % 3, 0이면 3번
            // 말한 횟수 : i / 3 + 1
            // 말한 단어 : words[i-1]
            if((i > 1 && words[i-1].charAt(0) != words[i-2].charAt(words[i-2].length()-1)) || set.contains(words[i-1])) {
                answer[0] = i%n == 0 ? n : i%n;
                answer[1] = (i-1)/n+1;
                return answer;
            }
            set.add(words[i-1]);
        }
        answer[0] = 0;
        answer[1] = 0;

        return answer;
    }
}
```
