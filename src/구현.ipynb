```java
// 프린터 큐(1966)

import java.util.*;
import java.io.*;

public class Main {

    // queue의 front 문서의 중요도가 1등이 아니면 맨뒤로
    // 특정 문서가 몇 번째로 인쇄되는지
    // 1(0) 1(1) 8(2) 5(3) 9(4) 2(5)
    // 9(4) 8(2) 5(3) 2(5) 1(0) 1(1)

    static class Document implements Comparable<Document> {
        int id;
        int weight;
        public Document(int id, int weight) {
            this.id = id;
            this.weight = weight;
        }
        @Override
        public int compareTo(Document o) {
            return Integer.compare(o.weight, this.weight);
        }
    }

    public static void main(String[] args) throws Exception {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        BufferedWriter bw = new BufferedWriter(new OutputStreamWriter(System.out));
        int T = Integer.parseInt(br.readLine());
        for(int t = 0; t < T; t++) {
            StringTokenizer st = new StringTokenizer(br.readLine());
            int N = Integer.parseInt(st.nextToken()); // N개의 문서
            int M = Integer.parseInt(st.nextToken()); // M번 문서가 궁금해
            PriorityQueue<Document> pq = new PriorityQueue<>();
            st = new StringTokenizer(br.readLine());
            Queue<Document> queue = new LinkedList<>(); // 문서 번호
            for(int i = 0; i < N; i++) {
                Document d = new Document(i, Integer.parseInt(st.nextToken()));
                pq.add(d);
                queue.add(d);
            }
            int order = 0;
            while(!queue.isEmpty()) {
                Document document = queue.remove();
                // 꺼낸게 우선순위 1등이면 꺼내
                if(pq.peek().weight == document.weight) {
                    order++;
                    if(document.id == M) {
                        bw.write(String.valueOf(order));
                        bw.newLine();
                        break;
                    }
                    pq.remove();
                }
                else { // 아니면 다시 넣어
                    queue.add(document);
                }
            }
        }
        bw.flush();
    }
}
```

```java
// 통계학(2108)

import java.util.*;
import java.io.*;

public class Main {

    public static void main(String[] args) throws Exception {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        BufferedWriter bw = new BufferedWriter(new OutputStreamWriter(System.out));
        int N = Integer.parseInt(br.readLine());
        int[] arr = new int[N];
        long sum = 0;
        HashMap<Integer, Integer> map = new HashMap<>();
        for(int i = 0; i < N; i++) {
            int num = Integer.parseInt(br.readLine());
            sum += (long)num;
            arr[i] = num;
            map.put(num, map.getOrDefault(num, 0) + 1);
        }
        Arrays.sort(arr);
        bw.write(String.valueOf(Math.round((double)sum/N)));
        bw.newLine();
        bw.write(String.valueOf(arr[N/2]));
        bw.newLine();
        int maxVal = 0;
        for(int key : map.keySet()) maxVal = Math.max(maxVal, map.get(key));
        LinkedList<Integer> list = new LinkedList<>();
        for(int key : map.keySet()) {
            if(map.get(key) == maxVal) list.add(key);
        }
        if(list.size() > 1) {
            Collections.sort(list);
            bw.write(String.valueOf(list.get(1)));
        }
        else bw.write(String.valueOf(list.get(0)));
        bw.newLine();
        bw.write(String.valueOf(arr[N-1]-arr[0]));
        bw.flush();
    }
}
```

```java
// 치킨 배달(15686)

import java.util.*;
import java.io.*;

public class Main {

    static class Pos {
        int r, c;
        public Pos(int r, int c) {
            this.r = r;
            this.c = c;
        }
    }

    static class Dist implements Comparable<Dist> {
        Pos house, chicken;
        int dist;

        public Dist(Pos house, Pos chicken) {
            this.house = house;
            this.chicken = chicken;
            this.dist = Math.abs(house.r-chicken.r) + Math.abs(house.c-chicken.c);
        }

        @Override
        public int compareTo(Dist o) {
            if(this.dist == o.dist) return Integer.compare(this.chicken.r, o.chicken.r);
            return Integer.compare(this.dist, o.dist);
        }
    }

    // 집(A개) 리스트, 치킨집(B개) 리스트
    // 각 집마다 각 치킨집까지의 거리를 가까운순으로 PriorityQueue<>[]
    // 치킨집을 M개씩 선택해서 치킨거리 구하고 최솟값 구하기
    public static void main(String[] args) throws Exception {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        BufferedWriter bw = new BufferedWriter(new OutputStreamWriter(System.out));
        StringTokenizer st = new StringTokenizer(br.readLine());
        int N = Integer.parseInt(st.nextToken());
        int M = Integer.parseInt(st.nextToken());
        ArrayList<Pos> house = new ArrayList<>();
        ArrayList<Pos> chicken = new ArrayList<>();
        for(int i = 1; i <= N; i++) {
            st = new StringTokenizer(br.readLine());
            for(int j = 1; j <= N; j++) {
                int n = Integer.parseInt(st.nextToken());
                if(n == 1) house.add(new Pos(i, j));
                else if(n == 2) chicken.add(new Pos(i,j));
            }
        }
        PriorityQueue[] pq = new PriorityQueue[house.size()];
        int[][] chickenDist = new int[house.size()][chicken.size()];
        for(int i = 0; i < house.size(); i++) {
            pq[i] = new PriorityQueue<Dist>();
            for(int j = 0; j < chicken.size(); j++) {
                Dist dist = new Dist(house.get(i), chicken.get(j));
                pq[i].add(dist);
                chickenDist[i][j] = dist.dist;
            }
        }
        
        // M개씩 선택해서 도시최소 치킨거리 구하기

    }
}
```
